```python
#Dijkstra's function
#vertex priority queue Q
import heapq

def Dijkstra(Graph, source):
    dist = {}
    prev = {}
    Q = [(0, source)]
    for v in Graph:
        if v != source:
            dist[v] = float('inf')
            prev[v] = None
            heapq.heappush(Q, (float('inf'), v))
        else:
            dist[v] = 0
    while Q:
        current_dist, u = heapq.heappop(Q)

        # Skip if already processed with a shorter distance
        if current_dist > dist[u]:
            continue

        # Explore neighbors
        for v, weight in Graph[u].items():
            alt = dist[u] + weight
            if alt < dist[v]:
                dist[v] = alt
                prev[v] = u
                # Update priority in the queue (add with new priority)
                heapq.heappush(Q, (alt, v))

    return dist, prev
```

```python


graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'C': 3, 'D': 2, 'E': 3},
    'C': {'B': 1, 'D': 4, 'E': 5},
    'D': {},
    'E': {'D':1}
}

distances, predecessors = Dijkstra(graph, 'A')

print("Shortest distances:", distances)
print("Predecessors:", predecessors)
```
